```
using System;
using System.IO;
using System.Text;
using System.Text.Json;

Person brother = new Person("Liam");
Person mum = new Person("Carol");
Person dad = new Person("Stuart");
Person person = new Person("Jordan")
{
   Father = dad,
   Mother = mum
};
person.Siblings.Add(brother);

public class Person
{
    public string FirstName { get; set; }
    public Person Father { get; set; }
    public Person Mother { get; set; }
    public List<Person> Siblings { get; set; } = new List<Person>();

    public Person(string firstname)
    {
        FirstName = firstname;
    }
}
```

## Preserving reference cycles in Json

```
JsonSerializerOptions options_preserve = new JsonSerializerOptions()
{
    ReferenceHandler = System.Text.Json.Serialization.ReferenceHandler.Preserve,
    WriteIndented = true
};
string json = JsonSerializer.Serialize(person, options_preserve);
Console.WriteLine(json);
```

## Ignore reference cycles in Json

```
JsonSerializerOptions options_ignore = new JsonSerializerOptions()
{
    ReferenceHandler = System.Text.Json.Serialization.ReferenceHandler.IgnoreCycles,
    WriteIndented = true
};
string json = JsonSerializer.Serialize(person, options_ignore);
Console.WriteLine(json);
```
